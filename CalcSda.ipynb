```python
import math

# Calculator Functions

def add(a, b):
    return a + b

def subtract(a, b):
    return a - b

def multiply(a, b):
    return a * b

def divide(a, b):
    if b == 0:
        return "Invalid value for denominator, can't divide by 0!"
    return a / b

def power(base, exponent):
    return base ** exponent

def square(x):
    return x ** 2

def sqrt(x):
    if x < 0:
        return "Invalid input for square root, can't handle negative values!"
    return math.sqrt(x)


def test_add():
    results = add(1, 2)
    assert results == 3

def test_subtract():
    results = subtract(10, 7)
    assert results == 3

def test_multiply():
    results = multiply(2, 3)
    assert results == 6

def test_divide():
    results = divide(6, 2)
    assert results == 3

def test_divide_by_zero():
    results = divide(99, 0)
    assert type(results) == str

def test_power():
    results = power(3, 2)
    assert results == 9

    results = power(3, 3)
    assert results == 27

def test_square():
    results = square(3)
    assert results == 9

def test_sqrt():
    results = sqrt(9)
    assert results == 3

    results = sqrt(16)
    assert results == 4

# Run all tests
def main():
    test_add()
    test_subtract()
    test_multiply()
    test_divide()
    test_divide_by_zero()
    test_power()
    test_square()
    test_sqrt()

if __name__ == "__main__":
    main()
```
